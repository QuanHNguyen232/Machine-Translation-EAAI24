```python
lang_pair_1 = 'en-de-fr-it-es'
lang_pair_2 = 'en-de-fr-pt-ro'
dataset_1 = load_data(f'/content/gdrive/MyDrive/Colab Notebooks/eaai24/dataset/{lang_pair_1}.pkl')
dataset_2 = load_data(f'/content/gdrive/MyDrive/Colab Notebooks/eaai24/dataset/{lang_pair_2}.pkl')
```

```python
len(dataset_1), len(dataset_2)
```

```python
pairs_1 = [pair['en'] for pair in dataset_1]
pairs_2 = [pair['en'] for pair in dataset_2]
```

```python
pairs_2_enDict = {pair:i for i, pair in enumerate(pairs_2)}
```

```python
align_idx_list = []
for i, pair in enumerate(tqdm(pairs_1)):
  en_sent = pair
  align_idx = pairs_2_enDict.get(en_sent, -1)
  align_idx_list.append((i, align_idx))
```

```python
align_idx_list_unk_rmved = [(k, v) for k, v in align_idx_list if v!=-1]
len(align_idx_list_unk_rmved), align_idx_list_unk_rmved[:10]
```

```python
save_dataset = []
for k, v in tqdm(align_idx_list_unk_rmved):
  pair_1 = dataset_1[k]
  pair_2 = dataset_2[v]
  save_dataset.append({
      'en': pair_1['en'],
      'de': pair_1['de'],
      'fr': pair_1['fr'],
      'it': pair_1['it'],
      'es': pair_1['es'],
      'pt': pair_2['pt'],
      'ro': pair_2['ro']
  })
```

## Load data

```python
connect_tok = '#'
PAD_token = 0
SOS_token = 1
EOS_token = 2
# MIN_LENGTH = 2
# MAX_LENGTH = 50

class Lang:
    def __init__(self, name):
        self.name = name
        self.max_len = 0
        self.word2index = {}
        self.word2count = {}
        self.index2word = {0:"<PAD>", 1: "<SOS>", 2: "<EOS>"}
        self.n_words = 3  # Count SOS and EOS

    def addWord(self, word):
        if word not in self.word2index:
            self.word2index[word] = self.n_words
            self.word2count[word] = 1
            self.index2word[self.n_words] = word
            self.n_words += 1
        else:
            self.word2count[word] += 1
```

```python
def unicodeToAscii(s):  # https://stackoverflow.com/a/518232/2809427
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')

def normalizeString(s): # Lowercase, trim, and remove non-letter characters
    s = s.lower().strip()
    # s = unicodeToAscii(s)
    # s = re.sub(r"([.!?])", r" \1", s)
    # s = re.sub(r"[^a-zA-Z.!?]+", r" ", s)
    return s
```

```python
filename = '/content/gdrive/MyDrive/Colab Notebooks/eaai24/dataset/en-de-fr-it-es-pt-ro.pkl'
# filename = '/content/gdrive/MyDrive/Colab Notebooks/eaai24/dataset/EnDeFrItEsPtRo-76k-most5k.pkl'
dataset = load_data(filename)
dataset[2]
```

```python
# create list of languages
lang1 = 'en'
lang2 = 'de'
lang3 = 'fr'
lang4 = 'it'
lang5 = 'es'
lang6 = 'pt'
lang7 = 'ro'
langs = [lang1, lang2, lang3, lang4, lang5, lang6, lang7]
```

```python
# create list of language objects
lang_objs_1 = {}
for lang in langs:
  lang_objs_1[lang] = Lang(lang)
```

```python
# create list of spacy for each language
spacy_langs_1 = {'en': spacy_en, 'de': spacy_de, 'fr': spacy_fr, 'it': spacy_it, 'es': spacy_es, 'pt': spacy_pt, 'ro': spacy_ro}
```

```python
origin_dataset = {i:{lang: pair[lang] for lang in langs} for i, pair in enumerate(tqdm(dataset))}
pairs_1 = {i:{lang:normalizeString(pair[lang]) for lang in langs} for i, pair in enumerate(tqdm(dataset))}
len(pairs_1), pairs_1[0]
```

```python
for i in tqdm(range(len(pairs_1))):
  sents = pairs_1[i]
  pair = {}
  for lang in langs:
    words = [tok.text for tok in spacy_langs_1[lang].tokenizer(sents[lang])]
    pair[lang] = connect_tok.join(words)
    for word in words:
      lang_objs_1[lang].addWord(word)
      assert lang_objs_1[lang].name == lang, lang_objs_1[lang].name

  pairs_1[i] = pair
```

## Sort freq dict

```python
N_MOST_FREQ = 10000
```

```python
for lang, langobj in lang_objs_1.items():
  print(f'{langobj.name}: num_words = {len(langobj.word2count)} , {len(langobj.word2index)}')
```

```python
# print number of words appear >= 2 times
for lang, langobj in lang_objs_1.items():
  print(f'{langobj.name}: num_words = ', len([k for k, v in langobj.word2count.items() if v>=2]))
```

```python
sorted_freq_by_langs_1 = {}
for lang, langobj in lang_objs_1.items():
  tmp_freq = {k: v for k, v in sorted(langobj.word2count.items(), key=lambda x: x[1], reverse=True)}
  sorted_freq_by_langs_1[langobj.name] = tmp_freq
  assert lang == langobj.name
```

```python
limited_sorted_freq_by_langs_1 = {}
for lang in sorted_freq_by_langs_1.keys():
  tmp_freq = {k: v for i, (k, v) in enumerate(sorted_freq_by_langs_1[lang].items()) if i < N_MOST_FREQ}
  limited_sorted_freq_by_langs_1[lang] = tmp_freq
```

## Filter pairs

```python
def checkValid(sent: list, dictionary, threshold=1.0):
  valid = [int(word in dictionary) for word in sent]
  return sum(valid)/len(valid) >= threshold # ratio of word in dicts >= threshold
```

```python
THRESHOLD = 1.0

filtered_idx_list_1 = []
# for each pair of sentences
for i in tqdm(range(len(pairs_1))):
  pair = pairs_1[i]
  isValid_langs = []
  # for each language
  for lang in pair.keys():
    tokens = pair[lang].split(connect_tok)
    limited_sorted_freq = limited_sorted_freq_by_langs_1[lang]
    isValid = checkValid(tokens, limited_sorted_freq, threshold=THRESHOLD)
    isValid_langs.append(isValid)
  if all(isValid_langs): filtered_idx_list_1.append(i)

len(filtered_idx_list_1)
```
